헤더정보 때문에 웹사이트의 정보를 읽어오지 못할 경우 아래에 접속하여  
https://developers.whatismybrowser.com/useragents/explore/layout_engine_name/trident/  
"Mozilla/5.0 (compatible, MSIE 11, Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko"를 헤더로 넣어줌  
상단 line 6, 8번을 항상 입력해주는 것을 추천

```python
# 타겟 사이트 → https://www.genie.co.kr/chart/top200?ditc=D&ymd=20200623&hh=21&rtm=Y&pg=1 # 1~4까지
import requests
from bs4 import BeautifulSoup
import time

header = {"User-Agent": "Mozilla/5.0 (compatible, MSIE 11, Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko"}
url = "https://www.genie.co.kr/chart/top200?ditc=D&ymd=20200623&hh=21&rtm=Y&pg="
genie = []

for i in range(1, 5):
    targetSite = "{}{}".format(url, i)
    request = requests.get(targetSite, headers = header)
    html = request.text
    soup = BeautifulSoup(html, "html.parser")

    # 노래, 가수 크롤링
    titles = soup.findAll("a", {"class": "title ellipsis"})
    artists = soup.findAll("a", {"class": "artist ellipsis"})

    time.sleep(1)
    
    # for title in titles:
    #     print(title.text.strip().split("\n")[-1].strip()) # 19금 노래가 \n으로 구분이 되어 구분자로 나누며, 공백은 split()으로 다시한번 없애준다.
    # for artist in artists[5:]: # 지니뮤직에서 5위까지의 순위를 계속 출력하기에 시작점을 5부터 진행한다.
    #    print(artist.text.strip().split("&")[0])
    for i in range(len(titles)):
        genie.append([titles[i].text.strip().split("\n")[-1].strip(), artists[i].text.strip().split("&")[0]])
```

```python
from datetime import datetime as dt

print(dt.now(), "현재")

for i in range(len(genie)):
    print("{0:3d}위. {1} - {2}".format(i + 1, genie[i][0], genie[i][1]))
```

```python
file = open("genieTop100.txt", "w", -1, "UTF-8")

file.write(str(dt.now()) + "\n")

for i in range(len(genie)):
    data = "{0:3d}위. {1} - {2}".format(i + 1, genie[i][0], genie[i][1]) + "\n"
    file.write(data)

file.close()
print("genieTop100.txt 파일에 쓰기 완료")
```
