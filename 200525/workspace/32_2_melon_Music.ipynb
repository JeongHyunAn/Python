헤더정보 때문에 웹사이트의 정보를 읽어오지 못할 경우 아래에 접속하여  
https://developers.whatismybrowser.com/useragents/explore/layout_engine_name/trident/  
"Mozilla/5.0 (compatible, MSIE 11, Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko"를 헤더로 넣어줌  
상단 line 6, 8번을 항상 입력해주는 것을 추천

```python
# 타겟 사이트 → https://www.melon.com/chart/index.htm
import requests
from bs4 import BeautifulSoup

targetSite = "https://www.melon.com/chart/index.htm"
header = {"User-Agent": "Mozilla/5.0 (compatible, MSIE 11, Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko"}

request = requests.get(targetSite, headers = header)
print(request) # <Response [406]> → 헤더 정보 때문에 읽어오지 못함
html = request.text
# print(html)
soup = BeautifulSoup(html, "html.parser")
print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll("div", {"class": "ellipsis rank01"})

# print(titles)

for title in titles:
    print(title.text.strip())
```

```python
# 가수 이름 크롤링
artists = soup.findAll("span", {"class": "checkEllipsis"})
# print(artists)

for artist in artists:
    print(artist.text.strip().split(",")[0])
```

```python
melon = [] # 노래 제목과 가수를 하나의 리스트로 만들어 저장

titles = soup.findAll("div", {"class": "ellipsis rank01"})
artists = soup.findAll("span", {"class": "checkEllipsis"})

for i in range(len(titles)):
    melon.append([titles[i].text.strip(), artists[i].text.strip().split(",")[0]])
    
from datetime import datetime as dt

print(dt.now(), "현재")

for i in range(len(melon)):
    print("{0:3d}위. {1} - {2}".format(i + 1, melon[i][0], melon[i][1]))
```

```python
file = open("melonTop100.txt", "w", -1, "UTF-8")

file.write(str(dt.now()) + "\n")

for i in range(len(melon)):
    data = "{0:3d}위. {1} - {2}".format(i + 1, melon[i][0], melon[i][1]) + "\n"
    file.write(data)

file.close()
print("melonTop100.txt 파일에 쓰기 완료")
```

```python
file = open("melonTop100.txt", "r", -1, "UTF-8")
lines = file.readlines()

for line in lines:
    print(line.rstrip())
    
file.close()
```
