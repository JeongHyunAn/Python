```python
import requests
from bs4 import BeautifulSoup
```

```python
# 타겟사이트 → http://dowellcomputer.com/main.jsp
request = requests.get("http://dowellcomputer.com/main.jsp")
# print(request)
html = request.text
# print(html)

# bs4 모듈의 BeautifulSoup() 메소드로 html 문서를 파싱할 준비를 한다.
soup = BeautifulSoup(html, "html.parser")
print(soup)
```

웹 페이지에 있는 객체에 대한 Elements의 위치를 보고자 한다면  
개발자 모드의 좌상단에 있는 마우스 커서모양을 클릭 후 해당 객체를 클릭하거나,  
Ctrl + Shift + C를 누른 후 해당 객체의 위에 해당 객체를 선택하면 된다.

```python
# html 태그와 html 태그에 지정된 속성을 이용한 크롤링
# 최근 공지사항의 제목은 class 속성이 tail인 td 태그 내부에 있다.
# <td class="tail"><a href="./notice/noticeViewForm.jsp?noticeID=4"><b>자바 기초 프로그래밍 강좌를 완강했습니다.</b></a></td>
# findAll("태그이름", {"속성이름": "속성값"}) → 속성이 없으면 딕셔러니 생략 가능

# notices = soup.findAll("td", {"class": "tail"})
# print(notices)

#notices = soup.findAll("b")
# for notice in notices:
#    print(notice)

# 자식 선택자(">") 및 자손 선택자(" "/공백)를 이용한 크롤링
# #viewBox > table > tbody > tr:nth-child(3) > td:nth-child(2) > a > b
# select("css 선택자")를 활용
# notices = soup.select("td b") # 자손 선택자
# notices = soup.select("td>a>b") # 자식 선택자
# for notice in notices:
#     print(notice)
    
notices = soup.select("td>a")

for notice in notices:
    # has_attr() 메소드는 인수로 지정된 속성이 있으면 True, 없으면 False를 리턴
    # print(notice.has_attr("href"))
    # get() 메소드는 인수로 지정된 속성의 값을 얻어온다.
    # print(notice.get("href"))
    if notice.get("href").find("notice") >= 0:
        print(notice.text)
```
