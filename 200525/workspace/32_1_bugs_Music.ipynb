벅스 뮤직 top 100 크롤링

```python
# 타겟 사이트 → https://music.bugs.co.kr/chart
import requests
from bs4 import BeautifulSoup

targetSite = "https://music.bugs.co.kr/chart"

request = requests.get(targetSite)
# print(request)
html = request.text
# print(html)
soup = BeautifulSoup(html, "html.parser")
# print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll("p", {"class": "title"})

# print(titles)

for title in titles:
    print(title.text.strip())
```

```python
# 가수 이름 크롤링
artists = soup.findAll("p", {"class": "artist"})
# print(artists)

# for artist in artists:
#     print(artist.text.strip()) # → 중복되는 값(복수의 아티스트)이 있어서 부가값이 출력됨

for artist in artists:
    print(artist.text.strip().split("\n")[0]) # 리스트로 변환해 준 후 0번째 인덱싱만 출력
```

```python
bugs = [] # 노래 제목과 가수를 하나의 리스트로 만들어 저장

titles = soup.findAll("p", {"class": "title"})
artists = soup.findAll("p", {"class": "artist"})

for i in range(len(titles)):
    bugs.append([titles[i].text.strip(), artists[i].text.strip().split("\n")[0]])
    
# print(bugs)

from datetime import datetime as dt

print(dt.now(), "현재")

for i in range(len(bugs)):
    print("{0:3d}위. {1} - {2}".format(i + 1, bugs[i][0], bugs[i][1]))
```

```python
# 크롤링 한 데이터를 텍스트 파일로 저장
# file = open("bugsTop100.txt", "w")
# 크롤링 데이터를 입력시 위와 같이 하면 "'cp949' codec can't encode character '\xa0' in position 20: illegal multibyte sequence"이란 오류가 발생
# 아래와 같이 UTF-8(유니코드)로 인코딩 시켜서 저장해야한다.
file = open("bugsTop100.txt", "w", -1, "UTF-8")

file.write(str(dt.now()) + "\n")

for i in range(len(bugs)):
    data = "{0:3d}위. {1} - {2}".format(i + 1, bugs[i][0], bugs[i][1]) + "\n"
    file.write(data)

file.close()
print("bugsTop100.txt 파일에 쓰기 완료")
```

```python
file = open("bugsTop100.txt", "r", -1, "UTF-8")
lines = file.readlines()

for line in lines:
    print(line.rstrip())
    
file.close()
```
